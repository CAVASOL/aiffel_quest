```python
#open diabetes data
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
```

```python
#create dataframe
diabetes_df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
diabetes_df.head()
```

```python
#add target
diabetes_df['target'] = dataset.target
```

```python
diabetes_df.head()
```

```python
#검증데이터 분리 - open lib
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    cancer_df.drop('target', axis=1),
    cancer_df['target'],
    test_size=0.3,
    random_state=1004)
```

```python
#train 데이터 확인(X)
X_train.head()
```

```python
#학습데이터 확인(y)
y_train.head()
```

```python
#data size
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#의사결정나무(Decision Tree)
from sklearn.tree import DecisionTreeClassifier

#모델 선택
model = DecisionTreeClassifier()

#학습
model.fit(X_train, y_train)

#예측
pred = model.predict(X_test)
pred
```

```python
#평가 accuracy_score(실제값, 예측값)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
# ML - Regression 회귀
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    diabetes_df.drop('target', axis=1),
    diabetes_df['target'],
    test_size=0.3,
    random_state=1004)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#LinearRegression 선형회귀

from sklearn.linear_model import LinearRegression

#모델 선택
model = LinearRegression()

#학습
model.fit(X_train, y_train)

#예측
pred = model.predict(X_test)
pred
```

```python
#mean_squared_error(실제값, 예측값) MSE - 낮을 수록 성능이 좋다!

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, pred)
```
