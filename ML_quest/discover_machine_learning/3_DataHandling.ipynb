```python
# apply 예시 - 칼로리가 1300 보다 큰 값과 작은 값
def cal(x):
    if x >= 1300:
        return "ye"
    else:
        return "nah"

df['cal'].apply(cal)
```

```python
# apply를 활용해서 새로운 컬럼 생성하기
df['gain_weight'] = df['cal'].apply(cal)
df
```

```python
# lambda, apply 활용
df['hesitat'] = df['cal'].apply(lambda x: 'ignore' if x >= 1300 else 'take it')
df
```
