```python
df = df.set_index(keys=['menu'])
df
```

```python
#loc 슬라이스, 양념치킨에서 마늘치킨의 호수에서 칼로리까지
df.loc['양념치킨':'마늘치킨', 'size':'cal']
```

```python
#iloc 슬라이스, 양념치킨에서 마늘치킨의 호수에서 칼로리까지
df.iloc[1:4, 1:3]
```

```python
#reset_index 기본값 drop=False
df = df.reset_index()
df
```

```python
df
```

```python
#할인율 컬럼 추가 [값은 0.2]
df['promo'] = 0.2
df
```

```python
#할인가 컬럼 추가
df['promo_price'] = df['price'] * (1-df['promo'])
df
```

```python
#결측값으로 추가, 원산지 컬럼을 만들고 결측값(NaN)으로 대입
import numpy as np

df['origin'] = np.nan
df
```

```python
#리스트 형태로 데이터(행) 추가,
aiffel = ['아이펠치킨', 16000, '11호', 1200, 0.5, '국내산', 8000]
df.loc['new'] = aiffel
df
```

```python
#딕셔너리 형태로 데이터(행) 추가
flip = {'menu':'풀잎치킨', 'price':10000, 'size':'10호'}
df.loc[10] = flip
df
```

```python
#후라이드의 원산지를 '국내산'으로 변경, (loc 활용)
df.loc[0, 'origin'] = '국내산'
df
```

```python
#파닭과 닭강정의 원산지는 '브라질'로 변경, (iloc 활용)
df.iloc[4, -2] = '브라질'
df.iloc[5, -2] = '브라질'
df
```

```python
#replace 활용해서 '반반'을 '양념반후라이드반'으로 변경
df = df.replace('반반', '양념반후라이드반')
df
```

```python
#replace 활용해서 '아이펠치킨'을 '[인기]아이펠치킨'으로, '풀잎치킨'을 '[베스트]풀잎치킨'으로
df = df.replace('아이펠치킨', '[인기]아이펠치킨').replace('풀잎치킨', '[베스트]풀잎치킨')
df
```

```python
#replace, 딕셔너리 활용. '후라이드'를 '황금후라이드'로, '양념치킨'을 '불닭양념치킨'으로
d = {'후라이드':'황금후라이드', '양념치킨':'불닭양념치킨'}
df = df.replace(d)
df
```

```python
#숫자변경 10000을 9900으로
df = df.replace(10000, 9900)
df
```

```python
#new_modudak.csv로 저장
df.to_csv('new_modudak.csv')
```
